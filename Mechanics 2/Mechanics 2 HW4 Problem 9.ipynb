## (b) Plucked string in Example 13.1

For the convenience, we set $h=\frac{\pi^{2}}{8}$, $L=\pi$, $\omega_{1}=\pi$. Then $f(X)=\sin{x}-\frac{1}{9}\sin(3x)+\frac{1}{25}\sin(5x)-\cdots$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,np.pi,1000)
j = 1

# 1/2 f(x+vt)

for k in range(0,8):
    t = k*np.pi/4
    q = 0.5*(np.sin(x+t)-np.sin(3*x+3*t)/9+np.sin(5*x+5*t)/25-np.sin(7*x+7*t)/49+np.sin(9*x+9*t)/81)
    plt.subplot(2,4,j)
    plt.xlim(0,np.pi)
    plt.ylim(-1.2,1.2)
    plt.plot(x,q,'k-',label=k)
    plt.legend(loc='upper right')
    j += 1

plt.show()
```

```python
j = 1

# 1/2 f(x-vt)

for k in range(0,8):
    t = k*np.pi/4
    q = 0.5*(np.sin(x-t)-np.sin(3*x-3*t)/9+np.sin(5*x-5*t)/25-np.sin(7*x-7*t)/49+np.sin(9*x-9*t)/81)
    plt.subplot(2,4,j)
    plt.xlim(0,np.pi)
    plt.ylim(-1.2,1.2)
    plt.plot(x,q,'k-',label=k)
    plt.legend(loc='upper right')
    j += 1

plt.show()
```

```python
j = 1

# 1/2 f(x+vt) + 1/2 f(x-vt)

for k in range(0,8):
    t = k*np.pi/4
    q1 = 0.5*(np.sin(x+t)-np.sin(3*x+3*t)/9+np.sin(5*x+5*t)/25-np.sin(7*x+7*t)/49+np.sin(9*x+9*t)/81)
    q2 = 0.5*(np.sin(x-t)-np.sin(3*x-3*t)/9+np.sin(5*x-5*t)/25-np.sin(7*x-7*t)/49+np.sin(9*x-9*t)/81)
    q = q1 + q2
    plt.subplot(2,4,j)
    plt.xlim(0,np.pi)
    plt.ylim(-1.2,1.2)
    plt.plot(x,q,'k-',label=k)
    plt.legend(loc='upper right')
    j += 1

plt.show()
```

## (d) Slightly different initial condition

```python
x = np.linspace(0,8,1000)
j = 1

for k in range(0,20):
    t = k*np.pi/10
    q1 = (32/np.pi**2)*(1-np.cos(np.pi/8))*np.sin(np.pi*x/8)*np.cos(t)
    q2 = -(32/np.pi**2)*(1-np.cos(3*np.pi/8))*np.sin(3*np.pi*x/8)*np.cos(3*t)/(3**2)
    q3 = (32/np.pi**2)*(1-np.cos(5*np.pi/8))*np.sin(5*np.pi*x/8)*np.cos(5*t)/(5**2)
    q4 = -(32/np.pi**2)*(1-np.cos(7*np.pi/8))*np.sin(7*np.pi*x/8)*np.cos(7*t)/(7**2)
    q5 = (32/np.pi**2)*(1-np.cos(9*np.pi/8))*np.sin(9*np.pi*x/8)*np.cos(9*t)/(9**2)
    q = q1 + q2 + q3 + q4 + q5
    plt.subplot(4,5,j)
    plt.xlim(0,8)
    plt.ylim(-1.2,1.2)
    plt.plot(x,q,'k-',label=k)
    plt.legend(loc='upper right')
    j += 1

plt.show()
```
