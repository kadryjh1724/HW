## wave at $t=0$

For the convenience, we set $h=\frac{\pi^{2}}{8}$, $L=\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt
pi = 3.141592

x = np.linspace(0,pi,1000)
q = np.sin(x)-np.sin(3*x)/9+np.sin(5*x)/25

plt.plot(x,q)
plt.show()
```

```python
q = np.sin(x)-np.sin(3*x)/9+np.sin(5*x)/25-np.sin(7*x)/49

plt.plot(x,q)
plt.show()
```

As you can see, including fourth term does not have significant effect. Therefore we can conclude that first three terms represents the initial displacement well.

## Sketch the shape of string

Note that

$\omega_{r}=\frac{r\pi}{L}\sqrt{\frac{\tau}{\rho}}$

I will use fundamental frequency as a standard. For convenience, let $\tau=\rho$, so $\omega_{1}=1$, $\omega_{3}=3$, and so on.

```python
x = np.linspace(0,pi,1000)
j = 1

for k in range(0,8):
    t = k*pi/4
    q = np.sin(x)*np.cos(t)-np.sin(3*x)*np.cos(3*t)/9+np.sin(5*x)*np.cos(5*t)/25
    plt.subplot(2,4,j)
    plt.xlim(0,pi)
    plt.ylim(-1.2,1.2)
    plt.plot(x,q,'k-',label=k)
    plt.legend(loc='upper right')
    j += 1

plt.show()
```
