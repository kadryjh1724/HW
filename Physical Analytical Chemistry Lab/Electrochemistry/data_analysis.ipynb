## Expt 1. CV of Sulfuric acid

```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
```

### Truncated CV of Sulfuric acid

```python
exp01 = pd.read_csv('exp01.csv', sep=',')
exp01_data = (exp01[["V","I"]].dropna(how="any")).values
V_1 = -exp01_data[:,0]
I_1= exp01_data[:,1] * 1e6

plt.plot(V_1, I_1, 'r-')
plt.xlabel('V (V)')
plt.ylabel('I (uA)')
plt.ylim(-15,10)
plt.title('Expt. 1, CV of sulfuric acid')
plt.show()
```

```python
plt.plot(V_1, I_1, 'r-')
plt.xlabel('V (V)')
plt.ylabel('I (uA)')
plt.title('Expt. 1, CV of sulfuric acid')
plt.show()
```

## Expt 3. Quantitative & Qualitative analysis with CV

우선, 미지시료 후보 용액 3가지(PFC: Potassium ferrocyanide, RHA: Rutheniumhexaamine(II) trichloride, FDM : Ferrocene dimethanol)의 CV를 plot해 보자.

```python
exp03_1 = pd.read_csv('exp03_1.csv', sep=',')
exp03_1_PFC = (exp03_1[["PFC_V","PFC_I"]].dropna(how="any")).values
exp03_1_RHA = (exp03_1[["RHA_V","RHA_I"]].dropna(how="any")).values
exp03_1_FDM = (exp03_1[["FDM_V","FDM_I"]].dropna(how="any")).values

V_PFC = exp03_1_PFC[:,0]
I_PFC = exp03_1_PFC[:,1] * 1e6
V_RHA = exp03_1_RHA[:,0]
I_RHA = exp03_1_RHA[:,1] * 1e6
V_FDM = exp03_1_FDM[:,0]
I_FDM = exp03_1_FDM[:,1] * 1e6
```

```python
plt.plot(V_PFC, I_PFC, 'r-', label='PFC')
plt.plot(V_RHA, I_RHA, 'g-', label='RHA')
plt.plot(V_FDM, I_FDM, 'b-', label='FDM')
plt.plot(V_3, I_3, 'k-', label='unknown')

plt.xlabel('V (V)')
plt.ylabel('I (uA)')
plt.legend(loc='upper left')
plt.title('Expt. 3-1, CV of three candidates and unknown')
plt.show()
```

```python
print('PFC, E_pc = 0.173339844', '\n')
print('RHA, E_pc = ', V_RHA[np.argmin(I_RHA)], '\n')
print('FDM, E_pc = ', V_FDM[np.argmin(I_FDM)], '\n')
print('unknown, E_pc = ', V_3[np.argmin(I_3)], '\n')

print('PFC, E_pa = ', V_PFC[np.argmax(I_RHA)], '\n')
print('RHA, E_pa = ', V_RHA[np.argmax(I_RHA)], '\n')
print('FDM, E_pa = ', V_FDM[np.argmax(I_FDM)], '\n')
print('unknown, E_pa = ', V_3[np.argmax(I_3)], '\n')
```

### Serial dilution of unknown 

0.1M KCl의 cyclic voltammogram의 전류 값을 빼서 faradaic current만을 분리하였다.

```python
exp03_3 = pd.read_csv('exp03_3.csv', sep=',')
exp03_3_data1 = (exp03_3[["V_1","I_1"]].dropna(how="any")).values
exp03_3_data2 = (exp03_3[["V_2","I_2"]].dropna(how="any")).values
exp03_3_data3 = (exp03_3[["V_3","I_3"]].dropna(how="any")).values
exp03_3_data4 = (exp03_3[["V_4","I_4"]].dropna(how="any")).values

V_d1 = exp03_3_data1[:,0]
I_d1 = exp03_3_data1[:,1] * 1e6 - I_KCl
V_d2 = exp03_3_data2[:,0]
I_d2 = exp03_3_data2[:,1] * 1e6 - I_KCl
V_d3 = exp03_3_data3[:,0]
I_d3 = exp03_3_data3[:,1] * 1e6 - I_KCl
V_d4 = exp03_3_data4[:,0]
I_d4 = exp03_3_data4[:,1] * 1e6 - I_KCl

plt.plot(V_d1, I_d1, 'r-', label='0.05mM')
plt.plot(V_d2, I_d2, 'g-', label='0.1mM')
plt.plot(V_d3, I_d3, 'b-', label='0.5mM')
plt.plot(V_d4, I_d4, 'k-', label='1mM')
plt.xlabel('V (V)')
plt.ylabel('I (uA)')
plt.legend(loc='upper left')
plt.title('Expt. 3-2, CV of diluted unknown(corrected)')
plt.show()
```

```python
print('0.05mM, I_pa = ', np.max(I_d1), '\n')
print('0.1mM, I_pa = ', np.max(I_d2), '\n')
print('0.5mM, I_pa = ', np.max(I_d3), '\n')
print('1mM, I_pa = ', np.max(I_d4), '\n')
```

```python
I_pa = [np.max(I_d1), np.max(I_d2), np.max(I_d3), np.max(I_d4)]
conc = [0.05, 0.1, 0.5, 1]

fitting = np.polyfit(conc, I_pa, 1)
p = np.poly1d(fitting)
t = np.arange(0.05, 1, 0.01)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(conc, I_pa, marker='x', color='black', s=10)
plt.plot(t, p(t), 'r-')
plt.xlabel('concentration (mM)')
plt.ylabel('I_pa (uA)')
plt.title('I_pa versus concentration of unknown')
plt.show()
```

## Expt 4. Chronoamperometry

우선, 아무 처리도 하지 않고 그래프의 개형을 관찰한다.

```python
exp04 = pd.read_csv('exp04.csv', sep=',')
exp04_data = (exp04[["t1","V1","I1","t2","V2","I2"]].dropna(how="any")).values
t_c1 = exp04_data[:,0]
V_c1 = exp04_data[:,1]
I_c1 = exp04_data[:,2] * 1e6
t_c2 = exp04_data[:,3]
V_c2 = exp04_data[:,4]
I_c2 = exp04_data[:,5] * 1e6
```

```python
plt.plot(t_c1, V_c1, 'r--', label='KCl V')
plt.plot(t_c1, I_c1, 'r-', label='KCl I')
plt.xlabel('time (s)')
plt.ylabel('V (V), I (uA)')
plt.legend(loc = 'upper right')
plt.title('Expt 4. Chronoamperometry')
plt.show()
```

```python
plt.plot(t_c2, V_c2, 'g--', label='unknown V')
plt.plot(t_c2, I_c2, 'g-', label='unknown I')
plt.xlabel('time (s)')
plt.ylabel('V (V), I (uA)')
plt.legend(loc = 'upper right')
plt.title('Expt 4. Chronoamperometry')
plt.show()
```

### Cottrell plot

```python
t_cott1 = pow(t_c1, -0.5)
t_cott2 = pow(t_c2, -0.5)

pd1 = np.hstack((t_cott1.reshape(-1,1), I_c1.reshape(-1,1)))
pd2 = np.hstack((t_cott2.reshape(-1,1), I_c2.reshape(-1,1)))

t_cott1_pd = pd.DataFrame(pd1)
t_cott2_pd = pd.DataFrame(pd2)
t_cott1_pd.to_csv("t_cott1.csv", index=False)
t_cott2_pd.to_csv("t_cott2.csv", index=False)
```

이 사이에, manual하게 header를 t와 I로 바꿔 주고 t가 큰 값을 골라 잘라낸다.

```python
exp04_cott1 = pd.read_csv('t_cott1.csv', sep=',')
exp04_cott2 = pd.read_csv('t_cott2.csv', sep=',')
exp04_cott1_data = (exp04_cott1[["t","I"]].dropna(how="any")).values
exp04_cott2_data = (exp04_cott2[["t","I"]].dropna(how="any")).values

t_cott1_trunc = exp04_cott1_data[:,0]
I_c1_trunc = exp04_cott1_data[:,1]
t_cott2_trunc = exp04_cott2_data[:,0]
I_c2_trunc = exp04_cott2_data[:,1]

plt.plot(t_cott1_trunc, I_c1_trunc, 'r-', label='KCl')
plt.plot(t_cott2_trunc, I_c2_trunc, 'b-', label='unknown')
plt.legend(loc='upper left')
plt.xlabel('t^-1/2')
plt.ylabel('I(uA)')
plt.title('Expt 4. Cottrell plot of KCl and unknown')
```

```python
cott1 = np.polyfit(t_cott1_trunc, I_c1_trunc, 1)
cott2 = np.polyfit(t_cott2_trunc, I_c2_trunc, 1)
p1 = np.poly1d(cott1)
p2 = np.poly1d(cott2)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(t_cott1_trunc, I_c1_trunc, marker='x', color='black', s=3)
ax.scatter(t_cott2_trunc, I_c2_trunc, marker='^', color='black', s=3)
plt.plot(t_cott1_trunc, p1(t_cott1_trunc), 'r-', label='KCl')
plt.plot(t_cott2_trunc, p2(t_cott2_trunc), 'g-', label='unknown')
plt.xlabel('t^-1/2')
plt.ylabel('I_pa (uA)')
plt.legend(loc='upper right')
plt.title('Expt 4. Cottrell plot of KCl and unknown')
plt.show()
```

```python
print(p1)
print(p2)
```

따라서 미지시료의 Cottrell plot에서 기울기는 4.335가 된다.

## Expt 6. pH dependence of Benzoquinone CV

```python
exp06 = pd.read_csv('exp06.csv', sep=',')
exp06_data = (exp06[["V1","I1","V2","I2","V3","I3"]].dropna(how="any")).values
V_pH1 = exp06_data[:,0]
I_pH1 = exp06_data[:,1] * 1e6
V_pH2 = exp06_data[:,2]
I_pH2 = exp06_data[:,3] * 1e6
V_pH3 = exp06_data[:,4]
I_pH3 = exp06_data[:,5] * 1e6
```

```python
plt.plot(V_pH1, I_pH1, 'r-', label='pH=4.69')
plt.plot(V_pH2, I_pH2, 'g-', label='pH=6.86')
plt.plot(V_pH3, I_pH3, 'b-', label='pH=8.21')
plt.xlabel('V (V)')
plt.ylabel('I (uA)')
plt.legend(loc='upper left')
plt.title('Expt. 6, pH dependence of Benzoquinone CV')
plt.show()
```

```python
print('pH = 4.69, deltaE_p = ', V_pH1[np.argmax(I_pH1)]-V_pH1[np.argmin(I_pH1)], '\n')
print('pH = 6.86, deltaE_p = ', V_pH1[np.argmax(I_pH2)]-V_pH1[np.argmin(I_pH2)], '\n')
print('pH = 8.21, deltaE_p = ', V_pH1[np.argmax(I_pH3)]-V_pH1[np.argmin(I_pH3)], '\n')
```

$\Delta E_{p}$의 pH dependence를 linear fit하면

```python
deltaE_p = [V_pH1[np.argmax(I_pH1)]-V_pH1[np.argmin(I_pH1)], V_pH2[np.argmax(I_pH2)]-V_pH2[np.argmin(I_pH2)], V_pH3[np.argmax(I_pH3)]-V_pH3[np.argmin(I_pH3)]]
pH = [4.69, 6.86, 8.21]

fitting = np.polyfit(pH, deltaE_p, 1)
p = np.poly1d(fitting)
t = np.arange(4.69, 8.21, 0.01)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(pH, deltaE_p, marker='x', color='black', s=10)
plt.plot(t, p(t), 'r-')
plt.xlabel('pH')
plt.ylabel('deltaE_p (V)')
plt.title('deltaE_p versus pH')
plt.show()
```
