# STM data analysis

## I-z graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
from scipy.optimize import curve_fit as cf

Iz = pd.read_csv('Iz.csv', sep=',')

Iz_pure = (Iz[["z1","I1"]].dropna(how="any")).values
Iz_C06 = (Iz[["z2","I2"]].dropna(how="any")).values
Iz_C10 = (Iz[["z3","I3"]].dropna(how="any")).values
Iz_C12 = (Iz[["z4","I4"]].dropna(how="any")).values
Iz_C18 = (Iz[["z5","I5"]].dropna(how="any")).values
```

```python
fig = plt.figure()
ax = fig.add_subplot(2,3,1)
ax.scatter(Iz_pure[:,0]*1e10, Iz_pure[:,1]*1e9, marker='.', color='black', s = 10)
plt.title('I-z data for Pure Gold')
plt.xlabel('x(angstroms)')
plt.ylabel('I(nA)')
ax1 = fig.add_subplot(2,3,2)
ax1.scatter(Iz_C06[:,0]*1e10, Iz_C06[:,1]*1e8, marker='.', color='black', s = 10)
plt.title('I-z data for C06')
plt.xlabel('x(angstroms)')
plt.ylabel('I(nA)')
ax2 = fig.add_subplot(2,3,3)
ax2.scatter(Iz_C10[:,0]*1e10, Iz_C10[:,1]*1e8, marker='.', color='red', s = 10)
plt.title('I-z data for C10')
plt.xlabel('x(angstroms)')
plt.ylabel('I(nA)')
ax3 = fig.add_subplot(2,3,4)
ax3.scatter(Iz_C12[:,0]*1e10, Iz_C12[:,1]*1e8, marker='.', color='blue', s = 10)
plt.title('I-z data for C12')
plt.xlabel('x(angstroms)')
plt.ylabel('I(nA)')
ax4 = fig.add_subplot(2,3,5)
ax4.scatter(Iz_C18[:,0]*1e10, Iz_C18[:,1]*1e8, marker='.', color='green', s = 10)
plt.title('I-z data for C18')
plt.xlabel('x(angstroms)')
plt.ylabel('I(nA)')
plt.show()
```

```python
def f1(x,c,phi):
    return np.exp(-1.025 * np.sqrt(phi) * (x - c))

Iz_pure[:,0] = - Iz_pure[:,0]
Iz_C06[:,0] = - Iz_C06[:,0]
Iz_C10[:,0] = - Iz_C10[:,0]
Iz_C12[:,0] = - Iz_C12[:,0]
Iz_C18[:,0] = - Iz_C18[:,0]

popt_p, pcov_p = cf(f1,Iz_pure[:,0],Iz_pure[:,1])
popt_06, pcov_06 = cf(f1,Iz_C06[:,0],Iz_C06[:,1])
popt_10, pcov_10 = cf(f1,Iz_C10[:,0],Iz_C10[:,1])
popt_12, pcov_12 = cf(f1,Iz_C12[:,0],Iz_C12[:,1])
popt_18, pcov_18 = cf(f1,Iz_C18[:,0],Iz_C18[:,1])

print('---- Work functions ----\n')
print('Pure gold : {} (eV)\n'.format(round(popt_p[1],4)))
print('C06 : {} (eV)\n'.format(round(popt_06[1],4)))
print('C10 : {} (eV)\n'.format(round(popt_10[1],4)))
print('C12 : {} (eV)\n'.format(round(popt_12[1],4)))
print('C18 : {} (eV)\n'.format(round(popt_18[1],4)))
```

## I-V graph

```python
IV = pd.read_csv('IV.csv', sep=',')

IV_pure = (IV[["V1","I1"]].dropna(how="any")).values
IV_C06 = (IV[["V2","I2"]].dropna(how="any")).values
IV_C10 = (IV[["V2","I3"]].dropna(how="any")).values
IV_C12 = (IV[["V2","I4"]].dropna(how="any")).values
IV_C18 = (IV[["V2","I5"]].dropna(how="any")).values
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(IV_pure[:,0], IV_pure[:,1]*1e9, marker='.', color='black', s = 10, label='Pure Gold')
ax.scatter(IV_C06[:,0], IV_C06[:,1]*1e9, marker='.', color='blue', s = 10, label='C6')
ax.scatter(IV_C10[:,0], IV_C10[:,1]*1e9, marker='.', color='red', s = 10, label='C10')
ax.scatter(IV_C12[:,0], IV_C12[:,1]*1e9, marker='.', color='green', s = 10, label='C12')
ax.scatter(IV_C18[:,0], IV_C18[:,1]*1e9, marker='.', color='pink', s = 10, label='C18')
plt.legend(loc = 'upper left')
plt.xlabel('Bias (V)')
plt.ylabel('Tunneling Current (nA)')
plt.title('I-V plot of alkanethiols on Au(111)')
plt.show()
```
