# Problem 2 of HW 3 - Thornton & Marion, 6-16 (b)

$z=x^{3/2}$, $y(x)=\frac{8}{13^{3/2}-8}\left[\left(1+\frac{9}{4}x\right)^{3/2}-1\right]$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

def z(x):
    return pow(x, 1.5)

def y(x):
    return 8.0 * (pow((1+2.25*x), 1.5) - 1) / (pow(13,1.5) - 8)
```

```python
x = np.arange(0, 1.001, 0.001)
y_range = np.arange(0, 1.001, 0.001)
x_mesh, y_mesh = np.meshgrid(x, y_range)

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection='3d')

ax.contour3D(x_mesh, y_mesh, z(x_mesh), 200, cmap=plt.cm.rainbow)
ax.plot(x, y(x), z(x), 'k-', linewidth = '5', label='y=y(x)')

ax.view_init(45, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.title('Problem 2 (b) plot')
plt.show()
```
