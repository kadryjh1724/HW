# Problem 10

Equation is $\ddot{\theta}=-\frac{g}{b}\sin\theta+\frac{\omega^{2}a}{b}\cos(\theta-\omega t)$ where $\theta(0)=0.2$, $\dot{\theta}(0)=0$, $a=0.2$, $b=1$, $g=1$, $\omega=0.1$.

$x=\theta$, $y=\dot{x}$, $\dot{y}=-\frac{g}{b}\sin x+\frac{\omega^{2}a}{b}\cos(x-\omega t)$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

a = 0.2
b = 1
g = 1

def ftn(z,t):
    x = z[0]
    y = z[1]
    dxdt = y
    dydt = - (g / b) * np.sin(x) + (0.1 * 0.1 * a / b) * np.cos(x - 0.1 * t)
    dzdt = [dxdt, dydt]
    return dzdt
```

```python
z0 = [0.2, 0]
n = 10000
    
t = np.linspace(0, 1000, n)
x = np.empty_like(t)
y = np.empty_like(t)
x[0] = z0[0]
y[0] = z0[1]

for i in range (1,n):
    tspan = [t[i-1], t[i]]
    z = odeint(ftn, z0, tspan)
    x[i] = z[1][0]
    y[i] = z[1][1]
    z0 = z[1]

plt.plot(t,x,'k-',linewidth='1')
plt.xlabel('time')
plt.ylabel('angle')
plt.ylim(-0.5,0.5)
plt.title('Plot of theta-t')
plt.show()
```
