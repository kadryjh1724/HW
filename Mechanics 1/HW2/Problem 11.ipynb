### (a) Reproducing Figure 4-23.

In python, there is a very convenient tool to visualize chaotic phenomena. This is pynamical package.

Reference: Boeing, G. 2016. “Visual Analysis of Nonlinear Dynamical Systems: Chaos, Fractals, Self-Similarity and the Limits of Prediction.” Systems, 4 (4), 37. doi:10.3390/systems4040037

Let me import some packages.

```python
import pynamical
from pynamical import simulate, bifurcation_plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Now, the bifurcation diagram goes. Pynamical package has built-in function for drawing bifurcation diagram for logistic map: extremely useful!

```python
pops = simulate(num_gens = 100, rate_min = 2.8, rate_max = 4.0, num_rates = 1000, num_discard = 200, initial_pop = 0.1)
bifurcation_plot(pops, xmin = 2.8, xmax = 4.0)
```

### (b) Zoomed-in version of bifurcation diagram

Now we restrict the diagram in $\alpha\in[3.840,3.857]$, $x_{n}\in[0.44,0.55]$. We just change change the limit from the option. Range of $\alpha$ and $x_{n}$ can be adjusted by changing range of $x$ and $y$ in the code, respectively.

```python
pops = simulate(num_gens = 100, rate_min = 3.840, rate_max = 3.857, num_rates = 1000, num_discard = 200, initial_pop = 0.1)
bifurcation_plot(pops, xmin = 3.840, xmax = 3.857, ymin = 0.44, ymax = 0.55)
```

### (c) Thornton & Marion, Problem 4-11.

Consider $\alpha=3.82831$ and $x_{1}=0.51$. Following code is identical code with the previous problem, Problem 3.

```python
alpha = 3.82831
k = 1000

map1 = np.zeros(k)
map1[0] = 0.51

for i in range (1,k):
    map1[i] = ( alpha * map1[i-1] * (1 - map1[i-1]) ) % 1

n = np.arange(0,k,1)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(n, map1, marker='.', color='black', s = 5)

plt.xlabel('number of iteration')
plt.ylabel('x')
plt.title('Problem 11(c)')
plt.show()
```

We can observe 3-cycle with $x=0.16,\,0.52,\,0.96$ for the first 80 cycles. First next periodic cycle occurs for approximately 200th cycle, until about 270th cycle.
