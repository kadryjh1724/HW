Use the function described in Example 4.3, $x_{n+1}=\alpha x_{n}(1-x_{n}^{2})$ where $\alpha=2.5$. Consider two starting values of $x_{1}$ that are similar, $0.900\,000\,0$ and $0.900\,000\,1$. Make a plot of $x_{n}$ versus $n$ for the two starting values and determine the lowest value of $n$ for which the two values diverge by more than 30 \%.

### Simple memoization code

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 2.5

map1 = np.zeros(100)
map2 = np.zeros(100)
map1[0] = 0.9000000
map2[0] = 0.9000001

# Memoization

for i in range (1,100):
    map1[i] = alpha * map1[i-1] * (1 - map1[i-1] ** 2)
    map2[i] = alpha * map2[i-1] * (1 - map2[i-1] ** 2)
    if (abs(map1[i] - map2[i]) / map1[i] >= 0.3): # Compute the fractional difference
        break

# From here, clumsy code for the plot. not important
plot1 = np.zeros(i+1)
plot2 = np.zeros(i+1)
n = np.arange(0, i+1, 1)

for j in range (0,i+1):
    plot1[j] = map1[j]
    plot2[j] = map2[j]

plt.plot(n, plot1, 'k-', label = '0.900 000 0')
plt.plot(n, plot2, 'k--', label = '0.900 000 1')
plt.legend(loc = 'upper right')
plt.xlabel('number of iteration')
plt.title('Problem 3')
plt.show()
```

Since we used the index $i$ from 0 and the loop broke at $i=29$,  we conclude that the lowest value of $n$ is 30.
