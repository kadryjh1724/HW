### Problem 4. Thornton & Marion 4-17.

The tent map is represented by the following iterations:

$ x_{n+1} = 2\alpha x_{n},\;\;(0<x<1/2),\;\;x_{n+1}=2\alpha(1-x_{n})\;\;(1/2<x<1) $

where $0<\alpha<1$. Make a map up to 20 iterations for $\alpha=0.4$ and $0.7$ with $x_{1}=0.2$. Does it appear that either of the maps represent chaotic behavior?

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 2.5

map1 = np.zeros(20)
map2 = np.zeros(20)
map1[0] = 0.2
map2[0] = 0.2

# Memoization

for i in range (1,20):
    if (map1[i-1] < 0.5):
        map1[i] = 2 * 0.4 * map1[i-1]
    else:
        map1[i] = 2 * 0.4 * (1 - map1[i-1])
    if (map2[i-1] < 0.5):
        map2[i] = 2 * 0.7 * map2[i-1]
    else:
        map2[i] = 2 * 0.4 * (1 - map2[i-1])

n = np.arange(0,20,1)

plt.plot(n, map1, 'k-', label = 'alpha = 0.4')
plt.plot(n, map2, 'k--', label = 'alpha = 0.7')
plt.legend(loc = 'upper right')
plt.xlabel('number of iteration')
plt.title('Problem 4')
plt.show()
```
